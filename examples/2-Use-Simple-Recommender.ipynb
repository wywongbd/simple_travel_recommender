### Suggest using user's hashtag feed 

In this approach, we try to construct clusters using posts selected from the **user's feed**. Meaning, while estimating **P(X | Y)** and **P(Y | X)**, we only sample from posts from the user's feed.  

The advantage of this is that it is much more efficient (we are not **crawling** here, because we are directly querying using **InstagramAPI**), so we can compute the topic clusters on the run.

```python
locations = ['china', 'europe', 'singapore', 'japan', 'australia', 'malaysia', 'korea']

try:
    suggestions, posts = obj.suggest(locations, (0.01, 0.005))
except Exception as error:
    print("Error while computing suggestions!")
```

If we compare these clusters to the ones obtained in **1-Construct-Topic-Clusters.ipynb**, they seem to contain much less hashtags. This is because in the **crawling** approach, we are sampling posts from the entire population of posts, not a restricted subset.

### Suggest using saved clusters 

In this approach, we provide recommendations using the pre-computed clusters (computed in **1-Construct-Topic-Clusters.ipynb**)

```python
DIR = "./output/"

travel_cluster = Cluster('travel', None, None, DIR + 'travel')
print(travel_cluster.tags)
```

```python
location_cluster_dict = {}
locations = ['japan', 'taiwan', 'korea', 'china', 'singapore', 'australia', 'europe']

for location in locations:
    location_cluster_dict[location] = Cluster(location, None, None, DIR + location)
    print(location)
    print(location_cluster_dict[location].tags)
    print('---------------------------------------------------------------------------------------')
```

```python
suggestions, posts = obj.suggest_with_clusters(travel_cluster, location_cluster_dict, 2, 1)
```
